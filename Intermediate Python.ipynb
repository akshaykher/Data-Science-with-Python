## Dictories & Pandas

Dictionary
* Dictionary is a collection of key value pairs
* keys should be unique
* keys should be immutable objects i.e. objects that can't be changed after they are created
    * string, boolean, integers and floats are immutable
        * Ex: a = "pqr", id(a) = x, if we append 's' then a = 'pqrs' but id(a) not equal to x
    * list however is mutable as its contents can be changed after created.
        * Ex: list1 = [1,2,3] where id(list1) = a, if we append 4 in this list then list1 = [1.2.3.4] but id(list1) = a only

```python
# dictionary with key as country and value as capitals
europe = { 'spain':'madrid', 'france':'paris', 'germany':'berlin', 'norway':'oslo' }

# display value (capital) for the key 'spain'
print(europe['spain'])

# Remove norway
del(europe['norway'])

# Print out the keys in europe
print(europe.keys())
```

```python
# Dictionary of dictionaries (key = country, value = capital and population)
europe = { 'spain': { 'capital':'madrid', 'population':46.77 },
           'france': { 'capital':'paris', 'population':66.03 },
           'germany': { 'capital':'berlin', 'population':80.62 },
           'norway': { 'capital':'oslo', 'population':5.084 } }


# Print out the capital of France
print(europe['france']['capital'])

# Create sub-dictionary data
data = {'capital':'rome','population':59.83}

# Add data to europe under key 'italy'
europe['italy'] = data

# Print europe
print(europe)
```

Pandas
* Pandas package is a high level data manipulation tool built on top of numpy
* Pandas DataFrame are very useful when working with tables (rows and columns)
* 2D numpy arrays can also be used but they can only hold 1 data type whereas Pandas DataFrames can hold multiple data types (i.e. each column can have a different data type)
* Row labels are row names (or index). By default they are named as 0,1,2,....
* Column labels are column names

```python
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Import pandas as pd
import pandas as pd

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {'country':names,'drives_right':dr,'cars_per_cap':cpc}

# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict)

# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

cars
```

#### Indexing a dataframe

* Each column of Pandas DataFrame is a Pandas Series. So if we paste together a bunch of Series, we get a df.
* Pandas series is like numpy 1d array but they are labeled (i.e. have a name of the series) OR 1D labelled numpy array
* Indexing can be done by:
   * Square brackets (either column access (through column names) or row access (through row index), can't do both)
   * loc - indexing using row and column labels (names) {remeber, default row labels (names) are 0,1,2.....}
   * iloc - indexing using row and column indexes (positions)

```python
##### Square brackets - Column Access #####

# print the Hour column
print(cutting_tool["Hour"])

# print the data type of Hour column
print(type(cutting_tool["Hour"]))
```

```python
##### Square brackets - Column Access #####

# USE [[ ]] if we want to have a single column being displayed as df and not a series

# print the Hour column
print(cutting_tool[["Hour"]])

# print the data type of Hour column
print(type(cutting_tool[["Hour"]]))
```

```python
##### Square brackets - Row Access #####

cutting_tool[1:4]
```

```python
##### LOC #####

# Output is Pandas Series as we are selecting only a single column
# : denotes all rows
cutting_tool.loc[:,"Oil"]
```

```python
##### LOC #####

# Output is Pandas DF as we are using double square brackets
# : denotes all rows

cutting_tool.loc[:,["Oil"]]
```

```python
##### LOC #####

# Note 0,1,2,3 are default row names here. They also happen to be row indexes. 
print(cutting_tool.loc[[0,1,2,3],:])

# changing index of the dataframe
cutting_tool.index = list(string.ascii_lowercase)[0:len(cutting_tool)]

# display dataframe
print(cutting_tool)

# same output with actual index names
print(cutting_tool.loc[['a','b','c','d'],:])

# running same command gives error as row names are alphabets and not default numbers
# print(cutting_tool.loc[[0,1,2,3],:])

# Resetting dataframe
cutting_tool = pd.read_csv('cutting_tool.csv')
```
